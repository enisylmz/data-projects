### Basic,Line Plot , Histogram, Scatter

```python
import matplotlib.pyplot as plt 
#Basic Plot 
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x,y)
plt.title('Basic Plot')
plt.show()
```

```python
# Line Plot 
plt.plot([1, 2, 3], [10, 20, 30], label= 'Trend', linestyle= '--', marker='o')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

```python
# Bar Chart 

categories = ['A','B','C']
values = [10, 15, 7]
plt.bar(categories, values, color = 'Blue')
plt.title('Bar Chart')
plt.show()
```

```python
# Histogram 
data = [1, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color= 'green', edgecolor = 'black')
plt.title('Histogram')

plt.show()
```

```python
# Scatter Plot
x = [1, 2, 3, 4, 5]
y = [10, 12, 25, 30, 45]
plt.scatter(x, y, color= 'red')
plt.title('Scatter Plot')
plt.legend('Dataset1')
plt.show()
```

### SEABORN

```python
import seaborn as sns
import numpy as np 
data= np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap= 'coolwarm')
plt.title('HeatMap')
plt.show()
```

```python
### Hands on Exercise
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
sales = [100, 140, 230, 269, 299, 345, 435]
plt.plot(years, sales, label= 'Sales Trend', marker= 'o', color = 'Green')
plt.legend('dATA-SET')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.show()
```

```python
# https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
# CORRELATON MATRX HEATMAP
correlation_matrix = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width',
       ]].corr()
sns.heatmap(correlation_matrix, annot= True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
