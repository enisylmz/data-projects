### As we determined our independent variables we can start using Logistic Regression to predict left values

```python
subdf = df[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years', 'salary']].copy()
subdf
```

```python
salary_dummies = pd.get_dummies(subdf.salary, prefix="salary", dtype=int)
df_with_dummies = pd.concat([subdf, salary_dummies], axis='columns')
df_with_dummies.drop('salary', axis='columns', inplace=True)
df_with_dummies.head()
```

```python
X = df_with_dummies
y = df.left
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision Score: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
print("Classification Score:\n", classification_report(y_test, y_pred))
```
